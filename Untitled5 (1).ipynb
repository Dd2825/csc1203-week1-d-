```python
import math 

def calculate_area(radius):
    area = math.pi * (radius ** 2) 
    return round(area, 2) 

radius = float(input("Enter the radius of the circle: "))

area = calculate_area(radius) 

circumference = lambda r: round(2 * math.pi * r, 2) 

print("Area:", area)
print("Circumference:", circumference(radius))
```

```python
import re 

def sum_numbers(numbers):
    return sum(numbers)

text = input("Enter a string: ") 

numbers = re.findall(r'\d+' , text)
numbers = [int(num) for num in numbers]

totel_sum = sum_numbers(numbers)

double_sum = lambda x: x * 2 

print("Numbers found:", numbers) 
print("Sum:", totel_sum)
print("Doubled value:", double_sum(totel_sum))
```

```python
import math 

def square_numbers(numbers): 
    square = lambda x: x ** 2 
    return list(map(square, numbers))

user_input = input("Enter numbers separated by commas: ")

numbers = [float(num) for num in user_input.split(",")]

squares = square_numbers(numbers)

max_square = max(squares)

sqrt_sum = math.sqrt(sum(squares))

print("Squares:", squares)
print("Maximum square:", max_square)
print("Squares root of sum of squares:", sqrt_sum)
```

```python
import re 
import math 

password = input("Enter a password: ") 

has_upper = re.search(r'[A-Z]', password)
has_lower = re.search(r'[a-z]', password)
has_digit = re.search(r'\d', password)
                      
if has_upper and has_lower and has_digit:
    strength = "Strong Password"
else: 
    strength = "Week Password"

scores_input = input("Enter 5 scores separated by spaces: ") 
scores = [float(score) for score in scores_input.split()]

def calculate_average(scores):
    avg = lambda x: sum(x) / len(x)
    return avg(scores)

average = calculate_average(scores)

rounded_avg = math.ceil(average)

print(strength)
print("Rounded Average Score:", rounded_avg)
```
